```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

df = pd.read_csv('C:/Users/slerr/Desktop/Projects/Financial_analysis_bi_product/2013_data_profit.csv')
df.head(25)
```

```python
#Processing the data
encoder = OneHotEncoder()
categorical_features = df[['Product','Discount Band','Segment','Country']]
encoded_features = encoder.fit_transform(categorical_features).toarray()
encoded_feature_names = encoder.get_feature_names_out(['Product','Discount Band','Segment','Country'])
encoded_df = pd.DataFrame(encoded_features,columns=encoded_feature_names)
```

```python
#CONCATE THE ENCODED DATA TO THE DATA SET
numerical_features = df.drop(columns =['Product','Discount Band','Segment','Country'])
final_df = pd.concat([numerical_features,encoded_df],axis=1)
```

```python
X = final_df.drop(columns =['Profit'])
Y= final_df['Profit']
```

```python
#Spiliting the data into training and testing sets
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.2,random_state=42)
```

```python
#Train the Model
```

```python
model = RandomForestRegressor(n_estimators =100,random_state =42)
```

```python
model.fit(X_train, Y_train)

#Precit the next years profit
Predicted_profit = model.predict(X_test)
mse = mean_squared_error(Y_test,Predicted_profit)
print(f'Mean Squared Error : {mse}')


```
